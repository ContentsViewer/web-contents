```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz, freqs

```

```python

# フィルタの設計
N = 4  # フィルタの次数
Wn = 1  # 正規化カットオフ周波数 (0から1の範囲で指定)

b, a = butter(N, Wn, analog=True, btype='low')

print(b, a)

# 周波数応答の計算
w, h = freqs(b, a)
# w, h = freqz(b, a)

# print(w, h)

# ボード線図の描画
plt.figure()
# plt.subplot(2, 1, 1)
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title('Butterworth filter frequency response')
plt.xscale('log')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.grid()

# plt.subplot(2, 1, 2)
# plt.plot(w, np.angle(h), 'b')
# plt.xlabel('Frequency [radians / second]')
# plt.ylabel('Phase [radians]')
# plt.grid()
# plt.show()
```
